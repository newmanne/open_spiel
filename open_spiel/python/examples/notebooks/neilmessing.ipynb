```python
import itertools
from tqdm import tqdm
```

```python
iters = 0
best = 0
if len(songs) != len(phrase):
    raise

for p in tqdm(itertools.permutations(songs)):
    success = True
    for i, s in enumerate(p):
        if phrase[i] not in s:
            success = False
            break
    if success:
        print(p)
        break
    else:
        iters += 1
        best = max(best, i)
    
    if iters % 100_000_000 == 0:
        print(iters, best)
        
print("DONE")
```
