```python
import numpy as np
tiny_history = np.array([
    [1, 0, 1],
])

short_history = np.array([
    [1, 0, 1],
    [2, 1, 0], 
])

medium_history = np.array([
    [1, 0, 1],
    [2, 1, 0], 
    [3, 3, 1],
])

history = np.array([
    [1, 0, 1],
    [2, 1, 0], 
    [3, 3, 1],
    [1, 1, 1],
])



```

```python
import pyspiel
import open_spiel.python.games
game = pyspiel.load_game('python_clock_auction', dict(filename='jan12_regional_vs_national_base.json')) 
state = game.new_initial_state()
```

```python
history[-3:]
```

```python
for h in [tiny_history, short_history, medium_history, history]:
    values = np.zeros((len(state.auction_params.all_bids), 3))
    # print(h[-3:, np.newaxis].shape) # mem x num_products x 1
    # print(state.auction_params.all_bids[np.newaxis, :, :].shape) # 1 x bundles x num_products
    one_hot = np.all(h[-3:, np.newaxis] == state.auction_params.all_bids[np.newaxis, :, :], axis=2).astype(int).T
    values[:, -one_hot.shape[1]:] = one_hot

    print(values)
    print()
```
