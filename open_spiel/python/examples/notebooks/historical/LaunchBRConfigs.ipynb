```python
import glob
import os
from pathlib import Path
EXPERIMENT = 'feb2_large'
CONFIG_DIR = '/apps/open_spiel/notebooks/configs/feb4_LR'

configs = glob.glob(f'{CONFIG_DIR}/*.yml')

# TODO: This won't launch things in the copy (they still write to original location)
# TODO: THe scripts just overwrite each other...

for br_script in glob.glob(f"/shared/outputs/{EXPERIMENT}/*/br*.sh"):
    if 'checkpoint' not in br_script:
        continue
    d = Path(br_script).parent
    for config in configs:
        config_name = Path(config).stem
        if d.stem == config_name:
            continue
        with open(br_script, 'r') as old_script_file:
            old_script = old_script_file.readlines()
        new_script_name = f'br_script_{config_name}.sh'
        
        idx = None
        for i, line in enumerate(old_script):
            if line.startswith('CMD='):
                idx = i
                break
        old_script[idx] = old_script[idx].replace('-- ', f'-- --output_suffix _{config_name} --config {config} ')

        with open(os.path.join(d, new_script_name), 'w') as f:
            f.write('\n'.join(old_script))
        !cd {d} && sbatch {new_script_name}
```
